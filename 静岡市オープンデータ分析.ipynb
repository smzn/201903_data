```r
#葵区だけの町を選択
des_data <- csv_data[csv_data$区=="葵区",]
#総数で並び替え
des_data <- des_data[order(des_data$総数, decreasing = TRUE),]
des_data
```

```r
barplot(t(des_data[1:100,5:22]), legend = labels, col = rainbow(18), names.arg = des_data[1:100,3])
```

```r
#散布図作成
#列選択リストを作成
columnList <- c("総数", "X70.74歳","X75.79歳","X80.84歳","X85歳以上")
#葵区の総数と70歳以上の人数を選択
scatter <- csv_data[csv_data$区=="葵区",columnList,]
#70歳以上の人数の和を求める
scatter_y <- scatter[,2]+scatter[,3]+scatter[,4]+scatter[,5]
```

```r
#回帰直線を求める
result.lm <- lm(scatter_y ~ scatter[,1])
#回帰直線の内容を表示
summary(result.lm)
#プロット
plot(scatter[,1], scatter_y, xlab = "Population", ylab = "70 Over")
#回帰直線を追加
abline(result.lm)
#R値を追加
mtext(substitute(paste(R^2," = ", text),list(text = round(summary(result.lm)$adj.r.squared,digits = 5))),line=-1,col=2,adj=0)
```

```r
preaoi.house <- csv_data[csv_data$区=="葵区","世帯数"]
presuruga.house <- csv_data[csv_data$区=="駿河区","世帯数"]
```

```r
population45.data <- read.csv("population15_3.data.csv")
population45.data
```

```r
#bartlett検定
#帰無仮説：「k群間の分散は等しい」
#対立仮説：「k群間の分散には差がある」
bartlett.test(population45.data[,2],population45.data[,1])
```

```r
#スティール・ドゥワス(Steel-Dwass)検定
#帰無仮説：「 k群間は似ている」
#対立仮説：「 k群間は差がある」
source("http://aoki2.si.gunma-u.ac.jp/R/src/Steel-Dwass.R", encoding="euc-jp")
Steel.Dwass(population45.data[,2], population45.data[,1])
```

```r
#bonferroni検定
#帰無仮説：「 k群間は似ている」
#対立仮説：「 k群間は差がある」
pairwise.t.test(population45.data[,2],population45.data[,1], p.adjust.method="bonferroni")
```

```r
#凝集型階層（観測値からクラスタへまとめていく）
library(cluster)
cluster.agnes <- agnes(population15.data, stand = T)
plot(cluster.agnes, which = 2, main = "", xlab = "", ylab = "")
```

```r
#区分型階層(全体から区分していく）
cluster.diana <- agnes(population15.data, stand = T) 
plot(cluster.diana, which = 2, main = "", xlab = "", ylab = "")
```

```r
#区切りを入力
cluster.h <- hclust(dist(population15.data), method = "ward.D")
plot(cluster.h, hang=-1) 
rect.hclust(cluster.h, k = 3, border = "red") 
```

```r
#k-means法
cluster.kmeans = kmeans(dist(population45.data), 6)
cluster.kmeans$cluster
cluster.kmeans$size
cluster.kmeans$withinss
```

```r
clusplot(population45.data, cluster.kmeans$cluster, color=TRUE, shade=TRUE, labels=2, lines=0)
```

```r
population.data <- read.csv("population.csv")
population.data
```

```r
result.lm <- lm(age80over ~ house, data=population.data)
summary(result.lm)
plot(population.data$house, population.data$age80over)
abline(result.lm)
```

```r
predict.data <- data.frame(house = c(3000))
predict.data
predict <- predict(result.lm, predict.data, interval = "prediction")
predict
```

```r
result.lm <- lm(age80over ~ house + age0to19, data=population.data)
summary(result.lm)
```

```r
predict.data <- data.frame(house = 3000, age0to19 = 500)
predict.data
predict <- predict(result.lm, predict.data, interval = "prediction")
predict
```
